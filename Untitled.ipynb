```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

X_train_full, X_valid_full, y_train, y_valid = train_test_split(
    X,
    y,
    train_size=0.8,
    test_size=0.2,
    random_state=0
)

ordinal_encoder = OrdinalEncoder(
    handle_unknown='use_encoded_value',
    unknown_value=-1
)

X_train_full['Compound'] = ordinal_encoder.fit_transform(
    X_train_full[['Compound']]
)

X_valid_full['Compound'] = ordinal_encoder.transform(
    X_valid_full[['Compound']]
)

OH_encoder = OneHotEncoder(
    handle_unknown='ignore',
    sparse_output=False
)

OH_cols_train = OH_encoder.fit_transform(
    X_train_full[['Race']]
)

OH_cols_valid = OH_encoder.transform(
    X_valid_full[['Race']]
)

OH_cols_train = pd.DataFrame(
    OH_cols_train,
    columns=OH_encoder.get_feature_names_out(['Race']),
    index=X_train_full.index
)

OH_cols_valid = pd.DataFrame(
    OH_cols_valid,
    columns=OH_encoder.get_feature_names_out(['Race']),
    index=X_valid_full.index
)

X_train_num = X_train_full.drop('Race', axis=1)
X_valid_num = X_valid_full.drop('Race', axis=1)

X_train_final = pd.concat([X_train_num, OH_cols_train], axis=1)
X_valid_final = pd.concat([X_valid_num, OH_cols_valid], axis=1)

```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

model = XGBClassifier(
    random_state=0,
    eval_metric='logloss'
)

model.fit(X_train_final, y_train)
```

```python
y_prob = model.predict_proba(X_valid_final)[:, 1]

auc_score = roc_auc_score(y_valid, y_prob)

print("AUC Score:", auc_score)
```

```python
fpr, tpr, thresholds = roc_curve(y_valid, y_prob)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"AUC = {auc_score:.4f}")
plt.plot([0,1], [0,1], linestyle='--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()

plt.show()
```
